```python
from sklearn.preprocessing import LabelEncoder

categorical_features = ['대출기간', '주택소유상태', '대출목적']

for i in categorical_features:
    le = LabelEncoder()
    le=le.fit(train_x[i])
    train_x[i]=le.transform(train_x[i])

    for case in np.unique(test_x[i]):
        if case not in le.classes_:
            le.classes_ = np.append(le.classes_, case)
    test_x[i]=le.transform(test_x[i])

display(train_x.head(3))
display(test_x.head(3))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, cut_tree
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from google.colab import files
import io
```

```python
Tss = np.sum((EuropeanCountries_data_stdz-EuropeanCountries_data_stdz.mean(0))**2)
R_square = []
K = range(1,10)
for k in K:
    Model = KMeans(n_clusters=k)
    Model.fit(EuropeanCountries_data_stdz)
    Wss = Model.inertia_
    Bss = Tss-Wss
    R_square.append(Bss/Tss)
```

```python
plt.figure(figsize=(10,8))
plt.plot(K, R_square, 'bx-')
plt.xlabel('k')
plt.ylabel('R-square')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
pca_corr_=pca_corr.fit_transform(EuropeanCountries_stdz)[:,0:2]
PCA_result = pd.concat([train_x.iloc[:,:], pd.DataFrame(pca_corr_[train_df['대출등급'].notnull()], columns = ['PC1', 'PC2']), train_df['대출등급']], 1)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(PCA_result.loc[PCA_result.대출등급 == 'Communism', 'PC1'],
            PCA_result.loc[PCA_result.대출등급 == 'Communism', 'PC2'], c='red', label = 'Communism')
plt.scatter(PCA_result.loc[PCA_result.대출등급 == 'Democracy', 'PC1'],
            PCA_result.loc[PCA_result.대출등급 == 'Democracy', 'PC2'], c='blue', label = 'Democracy')
for i, txt in enumerate(list(train_df['대출등급'])):
    plt.annotate(txt, (PCA_result.loc[i,'PC1'], PCA_result.loc[i,'PC2']))
plt.legend()
plt.axhline(y=0, color='gray', linewidth=0.5, linestyle='--')
plt.axvline(x=0, color='gray', linewidth=0.5, linestyle='--')
plt.show()
```
